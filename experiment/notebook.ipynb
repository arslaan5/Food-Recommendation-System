```python
def format_keywords(keywords):
    # Capitalize the first letter of each word and remove spaces
    formatted_words = [keyword.title() for keyword in keywords]
    return [word.replace(" ", "") for word in formatted_words]
```

```python
df['course'] = df['course'].apply(format_keywords)
df['cuisine'] = df['cuisine'].apply(format_keywords)
df['keyword'] = df['keyword'].apply(format_keywords)
df['ingredients'] = df['ingredients'].apply(format_keywords)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.neighbors import NearestNeighbors
```

```python
# Load pre-trained model for embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Convert tag descriptions into embeddings
tag_embeddings = model.encode(df['tags'].astype(str).tolist())
```

```python
tag_embeddings.shape
```

```python
tag_embeddings[0]
```

```python
# Train KNN model
knn = NearestNeighbors(n_neighbors=3, metric='cosine')  # 3 nearest dishes
knn.fit(tag_embeddings)
```

```python
# Function to recommend similar dishes using KNN
def recommend_dishes_knn(dish_name, df, knn_model):
    # Encode the dish name into an embedding
    dish_embedding = model.encode([dish_name])

    # Find K nearest neighbors
    distances, indices = knn_model.kneighbors(dish_embedding, return_distance=True)

    # Get recommendations
    recommendations = df.iloc[indices[0]][['name', 'imgurl']]
    
    return recommendations
```

```python
# Example Usage
print(recommend_dishes_knn("western", df, knn))
```
